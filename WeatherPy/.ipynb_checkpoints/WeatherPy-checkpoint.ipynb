```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python

# Define url
url = "http://api.openweathermap.org/data/2.5/weather?"
city_name = "Atlanta"
print(f"{url}q={city_name}&appid={weather_api_key}")
#weather_city = requests.get(f"{url}q={city_name}&appid={weather_api_key}").json()
#print(json.dumps(weather_city, indent=4, sort_keys=True))
#city_name = cities[0]
```

```python
# Because everytime the number of cities are different, first get the number of cities
n = len(cities)

# Define url
url = "http://api.openweathermap.org/data/2.5/weather?"

#print(f"{url}q={city_name}&appid={weather_api_key}")
#weather_city = requests.get(f"{url}q={city_name}&appid={weather_api_key}").json()
#print(json.dumps(weather_city, indent=4, sort_keys=True))
#city_name = cities[0]

# Initilize lists for data to be stored 
City = []      
Cloudiness = []
Country = []   
Date = []     
Humidity = []
Lat = []       
Lng = []  
Max_Temp = []
Wind_Speed = []


print("Beginning Data Retrieval")
print("-----------------------------")  

# Initilize the city number 
i = 1

for city in cities:
    
    # Create search query, make request and store in json
    query = f"{url}q={city}&appid={weather_api_key}"
    response = requests.get(query)
    response_json = response.json()
    
    # For each city, extract all the information needed
    try:
        name = response_json["name"]
        City.append(response_json["name"])
        Cloudiness.append(response_json["clouds"]["all"])
        Country.append(response_json["sys"]["country"]) 
        Date.append(response_json["dt"])
        Humidity.append(response_json["main"]["humidity"])
        Lat.append(response_json["coord"]["lat"])
        Lng.append(response_json["coord"]["lon"])
        Max_Temp.append(response_json["main"]["temp_max"])
        Wind_Speed.append(response_json["wind"]["speed"])
        
        print(f"Processing Record {i} out of {n} cities | {name}")
        
        i = i + 1
        
    # If weather information is missing, then do the following
    except:
        # Append null values
        print(f"City number={i} name={name} not found. Skipping...")
        i = i + 1
        pass
    
print("-----------------------------")
print("Data Retrieval Complete")      
print("-----------------------------")
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
# Perform a linear regression on temperature vs. latitude
from scipy.stats import linregress

def do_linreg(x_values, y_values):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    regress_values = x_values * slope + intercept
    # Plot scatter plot
    plt.scatter(x_values,y_values)
    plt.plot(x_values,regress_values,"r-")
    
    return [slope, intercept, rvalue, pvalue, stderr]
```

```python
####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression
```

```python
weather_data_north = weather_data.loc[weather_data['Lat'] >= 0]
weather_data_north.head()

x = weather_data_north["Lat"]
y = weather_data_north["Max Temp"]

params = do_linreg(x,y)
slope = params[0]
intercept = params[1]
rvalue = params[2]

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(40,300),fontsize=15,color="red")

# Label plot
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

# Print r square value
print(f"R squared: {rvalue**2}")
print(f"There is a strong negative linear relationship between max temp and latitude North Hemisphere,\
 and the high value of R squared also showed this is a good fit.")
# Show plot
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
weather_data_south = weather_data.loc[weather_data['Lat'] <= 0]
weather_data_south.head()

x = weather_data_south["Lat"]
y = weather_data_south["Max Temp"]


params = do_linreg(x,y)
slope = params[0]
intercept = params[1]
rvalue = params[2]

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(-50,300),fontsize=15,color="red")

# Label plot
plt.xlabel('Latitude')
plt.ylabel('Max Temp')

# Print r square value
print(f"R squared: {rvalue**2}")

print(f"There is a strong postive linear relationship between max temp and latitude in South Hemisphere,\
 but R squared dropped, which means a large proportion of the variations are not explained by this model.")

# Show plot
plt.show()

```
